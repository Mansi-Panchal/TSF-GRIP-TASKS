#### Step 1 - Import Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets as ds
```

#### Step 2 - Loading Dataset  

```python
iris = ds.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)

df.head()
```

#### Step 3 - Applying the Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier as dtc

models = []

for pair_index, pair in enumerate([[0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]]):
    X, y = iris.data[:, pair] , iris.target
    model = dtc().fit(X, y)
    models.append(model)
```

#### Step 4 - Visualizing Decision Tree

```python
from sklearn.tree import plot_tree as pt

dt = dtc(max_depth = 5).fit(iris.data, iris.target)

plt.figure(figsize = (20,15))
pt(dt, rounded = True, filled = True)
plt.show()
```
